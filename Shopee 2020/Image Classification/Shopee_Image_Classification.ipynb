# Preparation for Machine/Deep Learning
* The data set is way to large to be processed all at one go, so we will use the `ImageDataGenerator()` from TensorFlow to process small batches of images, one at a time.
* a batch of images consists of 128 images -- explore the functionalities of `train_data_gen` below to find out more

```python
# The 1./255 is to convert from uint8 to float32 in range [0,1].
# neural networks work better with small values
# we use 80% of the data set for training, 20% for validating
image_generator = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255, validation_split=0.2)
```

```python
BATCH_SIZE = 128
IMG_HEIGHT = 150
IMG_WIDTH = 150
EPOCHS = 15
```

```python
train_data_gen = image_generator.flow_from_directory(directory=str(train_dir),
                                                     batch_size=BATCH_SIZE,
                                                     shuffle=True,
                                                     target_size=(IMG_HEIGHT, IMG_WIDTH),
                                                     class_mode='categorical',
                                                     classes = list(CLASS_NAMES), 
                                                     subset='training')
valid_data_gen = image_generator.flow_from_directory(directory=str(train_dir),
                                                     batch_size=BATCH_SIZE,
                                                     shuffle=True,
                                                     target_size=(IMG_HEIGHT, IMG_WIDTH),
                                                     class_mode='categorical',
                                                     classes = list(CLASS_NAMES),
                                                     subset='validation')
```

```python
# the images have been resized to 150 pixels by 150 pixels large, each pixel with an RGB value hence the 3rd dimension is 3
print(train_data_gen.image_shape)
print(valid_data_gen.image_shape)
```

```python
# input: 
# image_batch: a dataset of images
# label_batch: the corresponding labels of image_batch
# this function plots 25 images from the image_batch in a 5-by-5 plot with its labels
def show_batch(image_batch, label_batch):
  plt.figure(figsize=(10,10))
  for n in range(25):
      plt.subplot(5,5,n+1)
      plt.imshow(image_batch[n])
      plt.title(CLASS_NAMES[label_batch[n]==1][0].title())
      plt.axis('off')
```

```python
# lets visualise some of the images in the dataset
image_batch, label_batch = next(train_data_gen)
show_batch(image_batch, label_batch)
```

# Deep Learning

Let's try a simple neural network and see how the model performs
* First flatten the (150,150,3) images into vectors of length 150 * 150 \* 3
* Next, include a fully-connected layer
* Lastly, include another fully-connected layer
* A summary of the model is provided by `model.summary()`

```python
input_shape = (IMG_WIDTH, IMG_HEIGHT, 3)

model = Sequential([
    Flatten(input_shape=input_shape),
    Dense(10, activation='relu'),
    Dense(42, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

print(model.summary())
```

```python
X, y = next(train_data_gen)
print(y)
```

```python
# this will take a long time to run, it has hundreds of thousands of parameters to train
# it will run 10 passes of forward and backward propagation
model.fit(train_data_gen, epochs=10)
```

```python
model.predict()
```
