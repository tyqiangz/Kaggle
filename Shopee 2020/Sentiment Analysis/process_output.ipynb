```python
import pandas as pd
import numpy as np
import emoji
```

```python
out_df = pd.read_csv("submission_25-07-2020_23-02.csv")
out_df.head()
```

```python
out_df.rating = out_df.rating.replace({0:1, 1:2, 2:3, 3:4, 4:5})
out_df.head()
```

```python
bad_emojis = ['👎🏻','👎','😈','😡','😠','😤','🤬']
```

```python
bad_emoji_bool = np.array([])

for i in range(out_df.shape[0]):
    for bad_emoji in bad_emojis:
        if bad_emoji in out_df.emojis:
            bad_emoji_bool.append(True)
            break
    bad_emoji_bool.append(False)
```

```python
submission_df = out_df.loc[:, ["review_id", "rating"]]
submission_df.head()
```

```python
submission_df.to_csv("prediction_25-07-2020_23-20.csv", index=False)
```
