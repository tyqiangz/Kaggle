Thus we can't just determine if a character is an emoji or not. The following function `find_emojis` helps to sieve out all the emojis from a text.

```python
def find_emojis(text):
    '''
    :param text: a string
    :output emoji_list: a list of emojis
    '''
    emoji_list = []
    data = regex.findall(r'\X', text)
    for word in data:
        if any(char in emoji.UNICODE_EMOJI for char in word):
            emoji_list.append(word)

    return emoji_list

def delete_repeated(text):
    return re.sub(r'(\w)\1{2,}', r'\1', text)

def process_text(text):
    '''
    :param text: a string representing a document
    :output: in order of sequence,
        list of emojis in text;
        list of words, without emojis in text;
        list of words, without emojis and reptition of more than 2 consecutive characters in text;
        list of words, lowercase, without unicodes, emojis and stopwords in text;
        list of stemmed words, lowercase, without unicodes, emojis and stopwords in text
    '''
    
    emojis = []
    words = []
    stopped_words = []
    stemmed_words = []
    
    # emojis are also unicodes, so we remove them from the text first
    emojis = find_emojis(text)
        
    # Create p_stemmer of class PorterStemmer
    p_stemmer = PorterStemmer()
    
    tokenizer1 = RegexpTokenizer(r'\w+')
    
    # list of words
    words = tokenizer1.tokenize(text)
    
    # list of non-stopwords words
    stopped_words = [word.lower() for word in words if word.lower() not in eng_stopwords]
    
    # list of stemmed words from list of non-stopwords words
    for i in stopped_words:
        stemmed_words.append(p_stemmer.stem(i))
    
    return " ".join(emojis), " ".join(words), delete_repeated(" ".join(words)), " ".join(stopped_words), " ".join(stemmed_words)
```

```python
# initialise some features that I wish to explore
new_features = ['emojis', 'no_emo_uni text', 'no_stopwords text', 'stemmed text']

for feature in new_features:
    train_df[feature] = np.nan
    test_df[feature] = np.nan
```

## Check if the emojis give a good indication on the ratings

```python
%%time
emoji_ratings = pd.DataFrame(columns = ["emoji", "rating_1", "rating_2", "rating_3", "rating_4", "rating_5"])
emoji_ratings.apply(pd.to_numeric)

for i in range(train_processed_df.shape[0]):
    if not pd.isnull(train_processed_df.loc[i, "emojis"]):
        emojis = list(set(train_processed_df.loc[i, "emojis"].split(" ")))
        for emoji in emojis:
            if emoji not in set(emoji_ratings.emoji):
                rating = train_processed_df.loc[i, "rating"]
                emoji_ratings.loc[emoji_ratings.shape[0], :] = [emoji,0,0,0,0,0]
                emoji_ratings.loc[emoji_ratings.emoji == emoji, f"rating_{rating}"] = 1
            else:
                emoji_ratings.loc[emoji_ratings.emoji == emoji, f"rating_{rating}"] += 1
emoji_ratings
```

```python
emoji_ratings["count"] = emoji_ratings.rating_1 + emoji_ratings.rating_2 + emoji_ratings.rating_3 + emoji_ratings.rating_4 + emoji_ratings.rating_5
```

Give a score to the emojis, suppose the emoji has [A,B,C,D,E] number of ratings from 1 to 5, then we compute the score as

$$ \text{score} = \dfrac{-1*A -0.5*B +0*C +0.5*D + 1*E}{A+B+C+D+E}.$$

An emoji would have a score close to -1 if it has mostly rating 1 and a score close to 1 if it has mostly rating 5.

```python
emoji_ratings["score"] = -emoji_ratings.rating_1 -0.5*emoji_ratings.rating_2\
    +0.5*emoji_ratings.rating_4 + emoji_ratings.rating_5

for i in range(emoji_ratings.shape[0]):
    emoji_ratings.loc[i, "score"] = emoji_ratings.loc[i, "score"] / emoji_ratings.loc[i, "count"]
```

Have to ensure that the columns of ratings are of int/numeric type for the `.style.background_gradient()` function to work.

```python
emoji_ratings = emoji_ratings.astype({f'rating_{i+1}': int for i in range(5)})
emoji_ratings.dtypes
```

Some of the emoji only appeared sparingly, which isn't very useful, so we only keep those that have appear for at least 15 times.

```python
emoji_ratings = emoji_ratings.loc[emoji_ratings["count"] >= 15, :]
emoji_ratings
```

# Wordclouds of the reviews

```python
def plot_wordcloud(doc_list, title):
    words = ''
    if hasattr(doc_list, '__iter__'):
        doc_list = [doc for doc in doc_list if not pd.isnull(doc)]
        words = " ".join(doc_list)
        
        wordcloud = WordCloud(width = 3000, height = 3000, 
                              background_color ='white',
                              min_font_size = 10).generate(words)

        # plot the WordCloud image                        
        plt.figure(figsize = (8, 8), facecolor = None) 
        plt.imshow(wordcloud)
        plt.title(title)
        plt.axis("off") 
        plt.tight_layout(pad = 0) 

        plt.savefig(title + '.png', bbox_inches='tight')
        plt.show()
```

```python
def seed_all(seed_value):
    random.seed(seed_value) # Python
    np.random.seed(seed_value) # cpu vars
    torch.manual_seed(seed_value) # cpu vars
    
    if torch.cuda.is_available(): 
        torch.cuda.manual_seed(seed_value)
        torch.cuda.manual_seed_all(seed_value) # gpu vars
        torch.backends.cudnn.deterministic = True  # needed
        torch.backends.cudnn.benchmark = False
```

```python
class TransformersBaseTokenizer(BaseTokenizer):
    """Wrapper around PreTrainedTokenizer to be compatible with fast.ai"""
    def __init__(self, pretrained_tokenizer: PreTrainedTokenizer, model_type = 'bert', **kwargs):
        self._pretrained_tokenizer = pretrained_tokenizer
        self.max_seq_len = pretrained_tokenizer.max_len
        self.model_type = model_type

    def __call__(self, *args, **kwargs): 
        return self

    def tokenizer(self, t:str) -> List[str]:
        """Limits the maximum sequence length and add the spesial tokens"""
        CLS = self._pretrained_tokenizer.cls_token
        SEP = self._pretrained_tokenizer.sep_token
        if self.model_type in ['roberta']:
            tokens = self._pretrained_tokenizer.tokenize(t, add_prefix_space=True)[:self.max_seq_len - 2]
            tokens = [CLS] + tokens + [SEP]
        else:
            tokens = self._pretrained_tokenizer.tokenize(t)[:self.max_seq_len - 2]
            if self.model_type in ['xlnet']:
                tokens = tokens + [SEP] +  [CLS]
            else:
                tokens = [CLS] + tokens + [SEP]
        return tokens
```

```python
transformer_tokenizer = tokenizer_class.from_pretrained(pretrained_model_name)
transformer_base_tokenizer = TransformersBaseTokenizer(pretrained_tokenizer = transformer_tokenizer, model_type = model_type)
fastai_tokenizer = Tokenizer(tok_func = transformer_base_tokenizer, pre_rules=[], post_rules=[])
```

```python
class TransformersVocab(Vocab):
    def __init__(self, tokenizer: PreTrainedTokenizer):
        super(TransformersVocab, self).__init__(itos = [])
        self.tokenizer = tokenizer
    
    def numericalize(self, t:Collection[str]) -> List[int]:
        "Convert a list of tokens `t` to their ids."
        return self.tokenizer.convert_tokens_to_ids(t)
        #return self.tokenizer.encode(t)

    def textify(self, nums:Collection[int], sep=' ') -> List[str]:
        "Convert a list of `nums` to their tokens."
        nums = np.array(nums).tolist()
        return sep.join(self.tokenizer.convert_ids_to_tokens(nums)) if sep is not None else self.tokenizer.convert_ids_to_tokens(nums)
    
    def __getstate__(self):
        return {'itos':self.itos, 'tokenizer':self.tokenizer}

    def __setstate__(self, state:dict):
        self.itos = state['itos']
        self.tokenizer = state['tokenizer']
        self.stoi = collections.defaultdict(int,{v:k for k,v in enumerate(self.itos)})
```
